```r
library(ggmsa)
library(msa)
library(ggplot2)
```

```r
ndha <- readAAStringSet("MSA/ndhA.pep")
```

```r
ndha
```

```r
ndha.aln <- msa(ndha,method='Muscle')
```

```r
ndha.aln
```

```r
protein_sequences <- ndha.aln@unmasked
```

```r
options(repr.plot.width=6, repr.plot.height=4)
# TLLSYSVIPFD
ggmsa(protein_sequences, 116, 136, color = "Chemistry_AA", 
      font = "TimesNewRoman", char_width = 0.5, seq_name = TRUE )+
        geom_seqlogo(color = "Chemistry_AA")+
        theme(axis.text = element_text(size=14,color='black'))
```

```r
options(repr.plot.width=6, repr.plot.height=4)
# LFLFIPITTRW
ggmsa(protein_sequences, 329, 349, color = "Chemistry_AA", 
      font = "TimesNewRoman", char_width = 0.5, seq_name = TRUE )+
        geom_seqlogo(color = "Chemistry_AA")+
        theme(axis.text = element_text(size=14,color='black'))
```
